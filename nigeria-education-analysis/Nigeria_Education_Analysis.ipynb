```python
import pandas as pd

# Load the data
df = pd.read_csv('/Users/omatasie-amadi/Documents/data science and analytics  portfolio/nigeria education analysis/data/nigeria_education_data/nigeria_education_data.csv')

print(f"Dataset shape: {df.shape[0]} rows, {df.shape[1]} columns")
print("\nColumn names:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
df.head()
```

```python
# Clean and reshape the data
import numpy as np

# checking what we have
print("Indicators in dataset:")
print(df['Series Name'].unique())
print(f"\nTotal indicators: {df['Series Name'].nunique()}")
```

```python
# Reshape from wide to long format
# Drop the metadata rows at the bottom
df_clean = df[df['Country Name'].notna()].copy()

# Get only the year columns
year_columns = [col for col in df_clean.columns if '[YR' in col]
print(f"Years available: {len(year_columns)} years from 2000 to 2024")

# Reshape the data
df_long = df_clean.melt(
    id_vars=['Country Name', 'Country Code', 'Series Name', 'Series Code'],
    value_vars=year_columns,
    var_name='Year',
    value_name='Value'
)

# Clean the Year column - extract just the year number
df_long['Year'] = df_long['Year'].str.extract('(\d{4})').astype(int)

# Convert Value to numeric (replace '..' with NaN)
df_long['Value'] = pd.to_numeric(df_long['Value'], errors='coerce')

# Remove rows with no data
df_long = df_long.dropna(subset=['Value'])

print(f"\n✅ Data reshaped")
print(f"New shape: {df_long.shape}")
print("\nSample of reshaped data:")
df_long.head(10)
```

## Key Findings: Primary School Enrollment Trends (2000-2010)

### Enrollment by Gender
- **Male enrollment**: Remained relatively stable around 70%, with a notable spike to 77% in 2007
- **Combined enrollment**: Averaged around 65%, peaking at 71% in 2007 before declining
- **Female enrollment**: Declined from 59% (2000) to 57% (2010), hitting a low of 56% in 2009 and a high of 63% in 2006

### Critical Pattern: Female Enrollment Decline
The most concerning trend is that female enrollment actually *decreased* over the decade, from 59% to 57%. This means that while male enrollment remained stable, girls' access to education worsened. This represents a significant setback for gender equity in Nigerian education.

### The 2006-2007 Anomaly
All enrollment categories spiked dramatically in 2007. This could reflect:
- A government education initiative or policy change
- Data collection methodology changes
- Temporary enrollment drives
The sharp decline immediately after suggests the spike was not sustained

### What the Numbers Mean
With female enrollment at only 57% in 2010, this means **43% of Nigerian girls of primary school age were not enrolled** - representing millions of girls without access to basic education.

## Analysis: Gender Gap in Primary Enrollment

### Persistent 12-Point Gap
The gender gap remained remarkably consistent at approximately **12 percentage points** throughout most of the 2000-2010 period. This stability suggests structural, systemic barriers to girls' education rather than temporary factors.

### The 2006-2007 Fluctuation
- **2006**: Gap narrowed to 8 percentage points (the best year on record)
- **2007**: Gap surged to ~15 percentage points (the worst year on record)
This volatility around 2006-2007 mirrors the enrollment spike we observed, suggesting data quality issues or policy changes during this period that affected genders differently.

### Why a 12-Point Gap Matters
In a country of Nigeria's size, a 12-point gender gap represents approximately **2-3 million girls** who are in school-aged but not enrolled, compared to boys. This isn't just an education issue - it's an economic development issue with generational impacts.

### Comparison to Global Standards
- UNESCO's goal: Achieve gender parity (0-point gap) in primary education
- Sub-Saharan Africa average: ~5-7 point gap
- Nigeria's 12-point gap indicates the country is lagging behind regional peers

## Adult Literacy Rates: Progress with Persistent Inequality

### Significant Improvement Over Two Decades
Adult literacy improved substantially from 2003 to 2024:
- **Total literacy**: 54% → 70% (+16 percentage points)
- **Female literacy**: 43% → 62% (+19 percentage points)  
- **Male literacy**: 67% → 78% (+11 percentage points)

This represents meaningful progress - approximately **25 million more Nigerian adults** became literate over this period.

### Gender Gap Persists in Adults
Despite improvements, a **16-point gap** remains between male (78%) and female (62%) adult literacy. This gap is even wider than the 12-point gap we see in current primary enrollment, reflecting the accumulated impact of decades of lower girls' education access.

### The Generational Effect
Adult literacy rates today reflect educational access from 20-40 years ago. The fact that current primary enrollment gaps mirror historical patterns suggests that without intervention, gender gaps in adult literacy will persist for decades to come.

### What 62% Female Literacy Means
Nearly **4 out of 10 Nigerian women** ages 15+ cannot read or write. This has cascading effects on maternal health, children's education, economic participation, and women's empowerment.

## Pupil-Teacher Ratio: Classroom Overcrowding Challenge

### Ratio Trends
Nigeria's primary school pupil-teacher ratio:
- Started at **43:1 in 2000** (well above UNESCO's 40:1 recommended maximum)
- Improved to **38:1 by 2010** (below the recommended threshold)
- **Above the 40:1 threshold** during: 2000-2001 and 2006-2009

### Progress with Setbacks
While the overall trend shows improvement (43:1 → 38:1), the ratio exceeded recommended levels during 7 of the 11 years measured. The spike during 2006-2009 coincides with the enrollment surge we observed, suggesting that increased enrollment outpaced teacher recruitment.

### What High Ratios Mean for Learning
When classrooms have 40+ students per teacher:
- **Individual attention decreases**: Teachers cannot provide personalized support
- **Class management challenges**: More time spent on discipline, less on instruction
- **Quality suffers**: Larger classes correlate with lower test scores and higher dropout rates

### The Hidden Story
National averages mask regional disparities. Rural areas often face ratios of **60:1 or higher**, while urban areas may be closer to 30:1. The students most in need of quality education (rural, low-income) face the largest classes.

### Teacher Shortage Crisis
To maintain a 40:1 ratio with current enrollment, Nigeria would need approximately **400,000 primary school teachers**. Achieving a 30:1 ratio (better for learning outcomes) would require over **500,000 teachers** - highlighting a massive workforce development need.

## Summary: Nigeria's Education Access Challenge

### The Big Picture: Three Interconnected Crises

**1. Female Education Crisis**
- Female primary enrollment *declined* from 59% to 57% (2000-2010)
- 12-point gender gap persists across enrollment
- 16-point gender gap persists in adult literacy
- **Millions of girls are being left behind**

**2. Capacity and Quality Challenges**  
- Pupil-teacher ratios frequently exceed recommended levels
- When enrollment surged (2007), teacher supply didn't keep pace
- This creates a quality-access tradeoff: more students but worse learning conditions

**3. Data Infrastructure Gaps**
- Limited enrollment data after 2010
- Literacy measured only periodically
- **We can't improve what we don't measure consistently**

### Key Insights for Policy

**Investment Priority**: The data suggests Nigeria faces not just an *access* problem but a *sustained access* problem - enrollment spikes aren't maintained. This points to:
- Economic barriers (school fees, opportunity costs)
- Infrastructure gaps (distance to schools, safety)
- Cultural factors (especially affecting girls)

**Teacher Recruitment is Critical**: Expanding enrollment without expanding the teaching force creates overcrowded classrooms that undermine learning quality.

**Focus on Girls**: With female enrollment declining while male enrollment holds steady, targeted interventions for girls' education are essential.

### What This Analysis Reveals
Nigeria has made progress on adult literacy (+16 points in two decades), showing that sustained effort can work. However, current primary enrollment trends - especially for girls - suggest that progress is stalling. Without renewed focus on girls' education and teacher capacity, gender gaps will persist for another generation.

### Recommendations for Further Research
- Regional analysis (North-South disparities are significant)
- Correlation with economic indicators (poverty rates, child labor)
- Completion rates and learning outcomes (enrollment ≠ learning)
- Comparative analysis with peer countries (Ghana, Kenya, Senegal)
